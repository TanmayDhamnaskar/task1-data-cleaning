```python
def modify_col(df):
    # Convert to clean format
    df.columns = [col.strip().lower().replace("-", "_") for col in df.columns]
    # Fix spelling mistakes
    df.rename(columns={"handcap": "handicap", "hipertension": "hypertension", "patientid": "patient_id", "appointmentid": "appointment_id",
                       "scheduledday" : "scheduled_day", "appointmentday": "appointment_day"}, inplace=True)
    return df
```

```python
df = modify_col(df)
```

```python
df['gender'] = df['gender'].str.lower().str.strip()
df['no_show'] = df['no_show'].str.lower().str.strip()
df['neighbourhood'] = df['neighbourhood'].str.title().str.strip()
```

```python
df
```

```python
df["gender"]= df["gender"].replace({"f": "female", "m": "male"})
df
```

```python
df['scheduled_day'] = pd.to_datetime(df['scheduled_day'])
df['appointment_day'] = pd.to_datetime(df['appointment_day'])
df['patient_id'] = df['patient_id'].astype(str)
```

```python
df.info()
```

```python
df['scheduled_day'] = df['scheduled_day'].dt.tz_localize(None)
df['appointment_day'] = df['appointment_day'].dt.tz_localize(None)
```
